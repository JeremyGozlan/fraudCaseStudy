```python
'''Comport EDA and modellings '''
```

```python
df=pd.read_json('files/data.json')
```

```python
df.info()
```

```python
smalldf=df[['acct_type','payout_type','delivery_method','has_header','previous_payouts','channels','email_domain','user_type']]
smalldf['event_length']=df['event_end']-df['event_start']
smalldf.loc[(smalldf["acct_type"]=='fraudster_event')]
```

```python
smalldf2=df[['acct_type','payout_type','delivery_method','has_header','previous_payouts','email_domain','channels','user_type']]
smalldf2['event_length']=df['event_end']-df['event_start']
smalldf2.loc[(smalldf2["acct_type"]=='premium')]
```

```python
df.acct_type.unique()
```

```python
df= df[df['user_type'].isin([1,3,4,5])]
```

```python
df['fraud'] = df['acct_type'].apply(lambda x: 1 if x in ['fraudster','fraudster_event','fraudster_att'] else 0)
```

```python
df['fraud']=df['fraud'].apply(lambda x: 1 if x==True else 0)
```

```python
df['previous_payouts']=df['previous_payouts'].apply(lambda x : 1 if len(x)>=1 else 0)
def quantitySold(x):
    lst = [i["quantity_sold"] for i in x]
    return sum(lst)
df["quantity_sold"] = df["ticket_types"].apply(quantitySold)
```

```python
df['generic_email']=df['email_domain'].apply(lambda x : 0 if  all(s not in x for s in ['aol', 'gmail', 'live','hotmail','yahoo','outlook']) else 1)
```

```python
df['delivery_method']=df['delivery_method'].apply(lambda x: 1 if x>=1.0 else 0)
```

```python
df=pd.get_dummies(df,columns=['user_type'])
```

```python
df=df[['generic_email','previous_payouts','delivery_method','sale_duration2','quantity_sold','user_type_1','user_type_3','user_type_3','user_type_5','fraud']]
```

```python
NotFraudCount=len(df.loc[df['fraud']==0])
FraudCount=len(df.loc[df['fraud']==1])
ToGo=NotFraudCount-FraudCount
```

```python
sampledDf=df[df['fraud']==1].sample(ToGo, replace=True)

finalDf = pd.concat([df,sampledDf],ignore_index=True)
NotFraudCount=len(finalDf.loc[finalDf['fraud']==0])
FraudCount=len(finalDf.loc[finalDf['fraud']==1])
print NotFraudCount,FraudCount
```

```python
y=finalDf.pop('fraud')
X=finalDf
```

```python
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.2, random_state=41)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_predicted=model.predict(X_test)
model.score(X_test,y_test)
```

```python
from sklearn.metrics import f1_score,precision_score,recall_score

f=f1_score(y_test, y_predicted, average='macro')
p=precision_score(y_test, y_predicted, average='macro')
r=recall_score(y_test, y_predicted, average='macro')
print f,p,r
```
