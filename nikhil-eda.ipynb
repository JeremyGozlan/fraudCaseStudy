```python
fraud = sub_df[sub_df["fraud"] == True]
not_fraud = sub_df[sub_df["fraud"] == False]
```

```python
%matplotlib inline

plt.figure(figsize = (10, 12))
plt.subplot(2, 1, 1)
fraud["num_order"].hist(bins = 100, color = "b", label = "fraud")
not_fraud["num_order"].hist(bins = 100, color = "g", label = "not fraud")
plt.xlim(-200, 500)
plt.title('num_order')
plt.legend()

plt.subplot(2, 1, 2)
fraud["has_logo"].hist(bins = 100, color = "b", label = "fraud")
not_fraud["has_logo"].hist(bins = 100, color = "g", label = "not fraud")
plt.xlim(-200, 500)
plt.title('has_logo')
plt.legend()
plt.show()

not_fraud["num_order"].count
```
